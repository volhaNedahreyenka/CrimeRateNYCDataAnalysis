```python
victims_by_race = crimedata['VIC_RACE'].value_counts(ascending=False)
victims_by_race
```

```python
plt.style.use("ggplot")
fig = plt.figure(figsize=(8,8))
victims_by_race.plot.barh(alpha=.6,width=0.5)
plt.xlabel("Number of victims", color='black')
plt.ylabel("Victim's race", color = 'black')
plt.title("Statistics on the victims' race in New York City Crimes", fontsize = 15, color='black') 
plt.savefig('victims_by_race.jpg')
plt.show()
```

```python
suspects_by_race = crimedata['SUSP_RACE'].value_counts(ascending=False)
suspects_by_race
```

```python
plt.style.use("ggplot")
fig = plt.figure(figsize=(10,10))
suspects_by_race.plot.barh(alpha=.6,width=0.5)
plt.xlabel("Number of suspects", color='black')
plt.ylabel("Suspect's race", color = 'black')
plt.title("Statistics on the suspects' race in New York City Crimes", fontsize = 15, color='black') 
plt.savefig('suspects_by_race.jpg')
plt.show()
```

```python
plt.figure(figsize=(7,7))
ax = sns.countplot('SUSP_SEX', data=crimedata, palette = "Set2")
ax.set_xlabel("Suspect's sex", color = 'black', fontsize=12)
ax.set_ylabel("Number of suspects", color = 'black', fontsize=12)
ax.set_title("Statistics on the suspects' sex in New York City Crimes", fontsize =14)
plt.savefig('suspectssex.jpg')
plt.show()
```

```python
from seaborn import axes_style

with axes_style({'axes.facecolor': 'bisque'}):
    plt.figure(figsize=(8,8))
    ax = sns.countplot('LAW_CAT_CD',data=crimedata,color='goldenrod',alpha=.6)  
    
ax.set_xlabel("Type of crime", color='black')
ax.set_ylabel("Frequency", color='black')
ax.set_title("Statistics on the type of crime in New York City", fontsize = 15) 
plt.savefig('typeofcrime1.jpg')
plt.show()
```

```python
plt.figure(figsize=(8,8))
ax = sns.countplot('BORO_NM', data=crimedata)
ax.set_xlabel("Name of the borough", color='black')
ax.set_ylabel("Number of crimes", color='black')
ax.set_title("Statistics on crimes in New York City by boroughs", fontsize=15) 
plt.savefig('boroughs.jpg')
plt.show()
```

```python
plt.figure(figsize=(14,9))
ax = sns.countplot(data=crimedata,x="SUSP_RACE", hue="SUSP_SEX")
ax.set_xlabel("Suspect's race", color='black')
ax.set_ylabel("Number of suspects", color = 'black')
ax.set_title("Statistics on suspects by their race and gender in New York City", fontsize = 15, color='black') 
plt.savefig('suspectsrace.jpg')
plt.show()
```

### Count the number of crimes by their location

```python
crime_loc = crimedata.groupby('PREM_TYP_DESC').size().sort_values(ascending=True).tail(20)
crime_loc
```

```python
plt.style.use("ggplot")
fig = plt.figure(figsize=(9,9))
crime_loc.plot.barh(alpha=.6,width=0.5)
plt.xlabel("Number of crimes", color='black')
plt.ylabel("Location", color = 'black')
plt.title("Statistics on crimes in New York City\n by location", fontsize = 15, color='black') 
plt.tight_layout()
plt.savefig('crimeslocations.jpg')
plt.show()
```

### Count the number of crimes by their type in police description

```python
crime_pd = crimedata.groupby('PD_DESC').size().sort_values(ascending=True).tail(10)
crime_pd
```

```python
plt.figure(figsize=(8,8))
crime_pd.plot.barh(alpha=.6,color='green',legend=False)
plt.xlabel("Number of crimes", color='black')
plt.ylabel("Types of crimes", color = 'black')
plt.title("The most frequent types of crimes\n in New York City", fontsize = 15, color='black') 
plt.tight_layout()
plt.savefig('mostfrequentcrimes.jpg')
plt.show()
```

```python
columns = ['PREM_TYP_DESC',###Specific description of premises; grocery store, residence, street, etc.
           'OFNS_DESC',    ###Description of offense corresponding with key code
           'Latitude',     ###Midblock Latitude coordinate for Global Coordinate System 
           'Longitude']    ###Midblock Longitude coordinate for Global Coordinate System
```

```python
columns_set = crimedata[columns]
columns_set.head()
```

```python
columns_set.isnull()
```

```python
columns_set.fillna(0, inplace=True) 
columns_set.head()
```

### Find all the rapes in the dataset.

```python
rapes = columns_set[columns_set["OFNS_DESC"] == "RAPE"]
rapes
```

### New York City map with the locations of rapes.

```python
map = folium.Map(location=(40.7128,-74.0060),zoom_start=10)

for each in rapes.iterrows():
    marker = folium.CircleMarker(location=[each[1]['Latitude'],each[1]['Longitude']],
                                          tooltip=each[1]['PREM_TYP_DESC'], radius=3,color='seagreen',
                                          fill_color='emerald')
    marker.add_to(map)
map.save('rapes_crimes_map.html')
display(map)
```

### Find all the crimes that happened on 12/25/2021

```python
christmasday = crimedata[crimedata['CMPLNT_FR_DT'] == "12/25/2021"]
christmasday
```

### Create a crime map for 12/25/2021

```python
map = folium.Map(location=(40.7128,-74.0060),tiles = 'Stamen Toner',zoom_start=10)

for each in christmasday.iterrows():
    marker = folium.Marker(location=[each[1]['Latitude'],each[1]['Longitude']],
                                          tooltip=each[1]['OFNS_DESC'])
    marker.add_to(map)
map.save('crimes_on_christmas_day_2021.html')
display(map)
```

### Find all the crimes that happened in 2019

```python
allCrimes2019 = crimedata[crimedata['CMPLNT_TO_DT'].dt.year==2019]
allCrimes2019
```

### Create a map for all the crimes that happened in 2019

```python
map = folium.Map(location=(40.7128,-74.0060),tiles = "Stamen Terrain",zoom_start=10)

for each in allCrimes2019.iterrows():
    marker = folium.CircleMarker(location=[each[1]['Latitude'],each[1]['Longitude']],
                                          tooltip=each[1]['OFNS_DESC'],radius=3,color='crimson',
                                          fill_color='crimson')
    marker.add_to(map)
map.save('crimes_in_2019.html')
display(map)
```
